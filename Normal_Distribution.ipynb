```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mu = 0  # mean
sigma = 1  # standard deviation

# Generate x values (observations)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

# Calculate the probability density function (PDF) for each x value
pdf = (1 / (sigma * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# Plotting the normal distribution chart
plt.plot(x, pdf)
plt.title("Normal Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values (observations)
x = np.linspace(-3, 3, 100)

# Calculate the probability density function (PDF) for each x value
pdf = (1 / np.sqrt(2*np.pi)) * np.exp(-0.5 * x**2)

# Plotting the standard normal distribution chart
plt.plot(x, pdf)
plt.title("Standard Normal Distribution (Z-distribution)")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```
